````python
def add_student(school, student, age, major):
  #Check that dictionary is a dictionary
  if not isinstance(school, dict):
    raise TypeError("School must be a dictionary")

  if not isinstance(student, str) or student.strip()=="":
    raise ValueError("Student must be a string and must not be empty")

  if not isinstance(age, int) or not 0 < age <= 25:
    raise ValueError("Age is not correct range")

  if not isinstance(major, str) or major.strip()=="":
    raise ValueError("Majot must be a non empty string")

  if student not in school:
    school[student]={"age":age, "major": major, "grades":{}}
    return
  #The student does exist but do not rely on dict structure

  if not isinstance(school[student],dict):
    raise TypeError(f"{student} value should be a dict but is corrupted")

  #Check all key value pairs we rely on
  if "age" not in school[student]:
    school[student]["age"]=age
  if "major" not in school[student]:
    school[student]["major"]=major
  if "grades" not in school[student]:
    school[student]["grades"]={}

  #ensure that grades is in the dictionary

  if not isinstance(school[student]["grades"],dict):
    raise ValueError(f"{student}'s grades value should be a dict")

  print(f"We just entered {school[student]}")




add_student(school, "Eliana", 19, "History")
````

#Q :: Return all letters that are unique

````python
#For letters that are seen exactly once

def exactly_once(items):
  answer= []
  unseen=set(items)
  for item in items:
    if item in unseen:
      unseen.remove(item)
      answer.append(item)
    else:
      if item in answer:
        answer.remove(item)
  return answer
print(exactly_once(["a","b","c","c","d","d"]))
````

#Q :: Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.


````python
def unique(items):
  a=set(items)
  if len(a)==items:
    return False
  else:
    return True
print(unique(["a","b","c","d","a"]))
````

````python
def containsDuplicate(nums):
    seen = set()
    for x in nums:
        if x in seen:
            return True
        seen.add(x)
    return False
````
